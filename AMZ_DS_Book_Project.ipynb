# <font color='orange'>**0) Steps to Convert Columns**</font>
Since my dataset contains some columns with mixed data type, I need to convert them before going any further .

```python
# Remove Non-Numeric Characters:
# Remove '%' from 'star5', 'star4', etc., and convert to numeric
# First, ensure all 'star' columns are treated as strings before replacing '%'
training_df['star5'] = training_df['star5'].astype(str).str.replace('%', '').astype(float) / 100
training_df['star4'] = training_df['star4'].astype(str).str.replace('%', '').astype(float) / 100
training_df['star3'] = training_df['star3'].astype(str).str.replace('%', '').astype(float) / 100
training_df['star2'] = training_df['star2'].astype(str).str.replace('%', '').astype(float) / 100
training_df['star1'] = training_df['star1'].astype(str).str.replace('%', '').astype(float) / 100

# For price ranges, take the average or first value
training_df['price (including used books)'] = training_df['price (including used books)'].astype(str).str.split('-').apply(lambda x: (float(x[0]) + float(x[-1])) / 2 if len(x) > 1 else float(x[0]))

# Convert 'price', 'n_reviews', and other columns to numeric
training_df['price'] = pd.to_numeric(training_df['price'], errors='coerce')
training_df['n_reviews'] = pd.to_numeric(training_df['n_reviews'], errors='coerce')

# Converting 'pages' to numeric
training_df['pages'] = pd.to_numeric(training_df['pages'], errors='coerce')

# check if all relevant columns are properly numeric
training_df.info()
```

## <font color="pink">**Outliers Analysis**</font>

### **Definition of Outlier**

 An outlier is a point or set of points that are different from other points. Sometimes they can be very high or very low.
Common methods to identify outliers include:
- **Statistical method**: Using Z-scores or the Interquartile Range (IQR) method. For the IQR method, a data point is considered an outlier if it falls below Q1 - 1.5 * IQR or above Q3 + 1.5 * IQR, where Q1 and Q3 are the first and third quartiles, respectively.

### **Handling Extreme Values**
If extreme values (outliers) are identified, several approaches can be taken:
1. **Examine the Outliers**: Investigate whether these outliers are valid entries (e.g., rare, expensive books) or data entry errors.
2. **Transformation**: Apply transformations (e.g., log transformation) to reduce the impact of outliers on statistical analyses.
3. **Removal**: If outliers are determined to be erroneous or irrelevant, they may be removed from the dataset.

---

### <font color="pink"> **Feature Selection**</font>

### **What is Feature Selection?**
Feature selection is the process of choosing the most important variables (features) from the dataset to improve model performance, avoid overfitting, and reduce computational costs.
#### **How Variance Helps in Feature Selection**:
- **Low-Variance Features**: Features with extremely low variance (like `star5`, `star4`, and other star ratings) provide little information because they don't change much across the dataset. These features can often be removed during feature selection, as they won't contribute significantly to the predictive power of a model.
- **High-Variance Features**: Columns like `price`, `n_reviews`, and `pages` have high variance, which means they might capture important differences in the data and could be useful for distinguishing between books when building models or making recommendations.
- **Techniques**: We can use methods like **Variance Threshold** (removing features with low variance) or **Recursive Feature Elimination (RFE)** to iteratively select important features.

## Removing outliers

**Goal:**

To determine the main categories or types of data science books in the dataset by analyzing their titles.

**Approach:**

We'll use Tf-idf (Term Frequency-Inverse Document Frequency) vectorization to convert text data (book titles) into numerical representations.
Once we have numeric representations, we will apply K-Means clustering to group book titles into clusters based on their semantic similarity.

# üí°**What are the main types of Data Science Book ?**

```python
clean_training_df.head(1)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

# Initialize the TfidfVectorizer with stopwords and ngram range
vectorizer = TfidfVectorizer(stop_words='english', ngram_range=(1, 2))

# Fit and transform the 'title' column of the clean_training_df DataFrame
X = vectorizer.fit_transform(clean_training_df['title'])

# Convert sparse matrix to a DataFrame
tfidf_df = pd.DataFrame(
    X.toarray(),
    columns=vectorizer.get_feature_names_out()
)

# Display the resulting DataFrame
tfidf_df.head()
```

```python
# Kmeans clustering
from sklearn.cluster import KMeans

# Initialize variables
sum_of_squared_distance = []
K = range(2, 10)

# Iterate over cluster sizes
for k in K:
    km = KMeans(n_clusters=k, max_iter=600, n_init=10, random_state=42)
    km.fit(X)  # Ensure X is defined, e.g., the TF-IDF matrix
    sum_of_squared_distance.append(km.inertia_)

# Print results or plot the Elbow Curve
print("Sum of Squared Distances:", sum_of_squared_distance)
```

```python
plt.plot (K, sum_of_squared_distance, 'bx-')
plt.xlabel("Number of clusters")
plt.ylabel('Sum of squared distance')
plt.title('Elbow Method For Optimal K')
plt.show()
```

```python
import plotly.graph_objects as go

# Create a Plotly figure
fig = go.Figure()

# Add a line plot for the Elbow Method
fig.add_trace(go.Scatter(
    x=list(K),  # Number of clusters
    y=sum_of_squared_distance,  # Sum of squared distances
    mode='lines+markers',
    line=dict(color='blue', dash='solid'),
    marker=dict(color='red', size=8),
    name='Inertia'
))

# Add axis labels and a title
fig.update_layout(
    title='Elbow Method For Optimal K',
    xaxis_title='Number of Clusters (k)',
    yaxis_title='Sum of Squared Distances (Inertia)',
    template='plotly_white'
)

# Show the plot
fig.show()
```

```python
# Number of clusters
true_k = 6

# Initialize the KMeans model
model = KMeans(n_clusters=true_k, init='k-means++', max_iter=600, n_init=10, random_state=42)

# Fit the model on the data (ensure X is the TF-IDF matrix or other feature matrix)
model.fit(X)

# Get the prediction labels
labels = model.labels_

# Create a DataFrame with book titles and their respective clusters
book_cl = pd.DataFrame(
    list(zip(clean_training_df['title'], labels)),
    columns=['title', 'cluster']
)

# Sort the DataFrame by cluster and display
book_cl.sort_values(by=['cluster'])
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Create word clouds for each cluster
for k in range(true_k):
    # Filter the titles belonging to cluster k
    text = " ".join(book_cl[book_cl['cluster'] == k]['title'])

    # Generate the word cloud
    wordcloud = WordCloud(max_font_size=50, max_words=100, background_color='white').generate(text)

    # Create subplots for the word clouds
    plt.subplot(2, 3, k + 1)  # Adjust grid dimensions if needed
    plt.title(f'Cluster {k}')
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')

# Display all subplots
plt.tight_layout()
plt.show()
```
